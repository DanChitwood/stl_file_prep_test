```python
import trimesh
import numpy as np
import matplotlib.pyplot as plt
```

# Mesh.stl

```python
# load in stl file
mesh = trimesh.load_mesh('mesh.stl')

# get vertices and faces
vertices = mesh.vertices
faces = mesh.faces

print("There are", len(vertices), "vertices and", len(faces), "faces")

#mesh.show()
```

```python
# set cropping limits
x_min = -0.13 # magenta
x_max = 0.17 # magenta
y_min = 0.33 # orange
y_max = 0.54 # orange
z_min = -0.14 # purple
z_max = 0.18 # purple
```

```python
# create scatter plot side view for cropping
plt.figure(figsize=(20,10))

pt_s = 0.001 # set point size

plt.subplot(1,2,1)
plt.scatter(vertices[:,0], vertices[:,1], s=pt_s)
plt.axvline(x_min, c="magenta")
plt.axvline(x_max, c="magenta")
plt.axhline(y_min, c="orange")
plt.axhline(y_max, c="orange")
plt.gca().set_aspect("equal")

plt.subplot(1,2,2)
plt.scatter(vertices[:,0], vertices[:,2], s=pt_s)
plt.axvline(x_min, c="magenta")
plt.axvline(x_max, c="magenta")
plt.axhline(z_min, c="purple")
plt.axhline(z_max, c="purple")
plt.gca().set_aspect("equal")

plt.tight_layout()
```

```python
# create minimum and maximum bounds of bounding box
min_bound = np.array([x_min, y_min, z_min]) 
max_bound = np.array([x_max, y_max, z_max])

# Get mask of vertices inside the bounding box
inside_mask = np.all((mesh.vertices >= min_bound) & (mesh.vertices <= max_bound), axis=1)

# Keep faces where all three vertices are inside the bounding box
face_mask = np.all(inside_mask[mesh.faces], axis=1)

# Create new mesh with only those faces
cropped_mesh = mesh.submesh([face_mask], append=True)

# get vertices and faces of cropped
cropped_vertices = cropped_mesh.vertices
cropped_faces = cropped_mesh.faces
```

```python
cropped_mesh.show()
```

```python
cropped_mesh.export('cropped_mesh.stl')
```
